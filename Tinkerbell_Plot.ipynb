```
import matplotlib.pyplot as plt

def x_coord(x,y):
  return (x ** 2) - (y ** 2) + (.9 * x) - (.6 * y)

def y_coord(x,y):
  return (2 * (x * y)) + (2 * x) + (.5 * y)

x_coordinate = [-0.7]
y_coordinate = [-0.6]

for i in range(1,100000):
  x = x_coord(x_coordinate[-1], y_coordinate[-1])
  y = y_coord(x_coordinate[-1], y_coordinate[-1])

  x_coordinate.append(x)
  y_coordinate.append(y)

plt.figure(figsize=(14,14))
plt.plot(x_coordinate, y_coordinate, 'o', ms="1")
plt.show()
```
